# Removing rows
Dropping all animals that unknown if they are endangered or not becauce lack of researches

```python
df.dropna(subset=['Is_Endangered'], inplace=True)
```

There are missing values in 'Class' 'Order' 'Family' during the lack of researches I had to remove them to get a better model

```python
df.dropna(subset=['Class', 'Order', 'Family', 'Diet'], inplace=True)
```

```python
df[['Class', 'Order', 'Family']] = df[['Class', 'Order', 'Family']].apply(lambda x: x.str.replace('†', '')) #removing symbol from text
```

```python
df
```

```python
df.info()
```

I filled missing numeric data with median

```python
df['Weight(kg)'] = df['Weight(kg)'].fillna(value=df['Weight(kg)'].mean())
df['Lifespan(years)'] = df['Lifespan(years)'].fillna(value=df['Lifespan(years)'].mean())
df['Length(cm)'] = df['Length(cm)'].fillna(value=df['Length(cm)'].mean())
```

```python
df
```

```python
df.info()
```

# Convert object types to categorical types

```python
df['Class'] = df['Class'].astype('category').cat.codes
df['Diet'] = df['Diet'].astype('category').cat.codes
# © https://stackoverflow.com/questions/28910851/python-pandas-changing-some-column-types-to-categories
```
